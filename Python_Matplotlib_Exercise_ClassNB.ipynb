```python
mean_value=csd['Shirt'].mean()
csd.fillna(mean_value, inplace=True)
```

## Exercise 1.Read Bags, Top, Shirt and Pant sales data and show it  using a multiline plot


```python
plt.plot(csd["month_number"], csd["Pant"],'-o' ,label="pant sales data")
plt.plot(csd["month_number"], csd["Shirt"],'-o' , label="shirt sales data")
plt.plot(csd["month_number"], csd["Bags"], '-o' ,label="bags sales data")
plt.plot(csd["month_number"], csd["Top"],'-o' , label="top sales data")


plt.title("Sales Data")
plt.xlabel("Month number")
plt.ylabel("Sales units in number")
plt.legend()
plt.show()
```

## Exercise 2: Read Pant and Shirt sales data of each month and show it using a scatter plot


```python
pant_sales = csd['Pant']
shirt_sales = csd['Shirt']
months = csd['month_number']
fig, ax = plt.subplots()
ax.scatter(months, pant_sales, label='Pant Sales Data')
ax.scatter(months, shirt_sales, label='Shirt Sales Data')
ax.set_xlabel('Month Number')
ax.set_ylabel('Number of units Sold')
ax.legend()
plt.show()
```

## Exercise 3: Read Jeans and Top product sales data and show it using the bar chart

Note:
-  move x axis of jeans .25 left and Top .25 right and add width
-  plt.bar(sales['month_number']-.25, sales['Jeans'],color='red',width=.30, label="Jeans sales data")

```python
jeans_sales = csd['Jeans']
top_sales = csd['Top']
months = csd['month_number']
fig, ax = plt.subplots()
ax.bar(months-0.25, jeans_sales, color='red', width=0.3, label='Jeans Sales Data')
ax.bar(months+0.25, top_sales, color='blue', width=0.3, label='Top Sales Data')
ax.set_xlabel('Month Number')
ax.set_ylabel('Sales units in number')
ax.legend()

plt.show()
```

## Exercise 4: Read the total profit of each month and show it using the histogram to see the most common profit ranges 
The histogram should look like this:


```python
total_profit = csd['total_profit']
fig, ax = plt.subplots()
ax.hist(total_profit)
ax.set_xlabel('Total Profit in Dollars')
ax.legend(['Profit Data'])
plt.show()
```

## Exercise 5: Calculate `total sale` data for last year for each product and show it using a Pie chart

- salesData=[sales['Bags'].sum(),.......]

```python
pant_total = csd['Pant'].sum()
shirt_total = csd['Shirt'].sum()
bags_total = csd['Bags'].sum()
top_total = csd['Top'].sum()
jean_total = csd['Jeans'].sum()
tie_total = csd['Tie'].sum()
fig, ax = plt.subplots()
ax.pie([pant_total, shirt_total, bags_total, top_total, jean_total, tie_total], 
       labels=['Pants', 'Shirts', 'Bags', 'Tops', 'Jeans', 'Tie'], autopct='%.1f%%')
ax.set_title("Sales data")
plt.show()
```

### Exercise 6: Read all product sales data and show it using the stack plot. Save the result as `stack_sales.png`


```python
x=csd['month_number']
y=np.vstack([csd['Pant'], 
             csd['Shirt'],
            csd['Bags'],
            csd['Top'],
            csd['Jeans'], 
            csd['Tie']])
```

```python
labels=['Pant', 'Shirt', 'Bags',  'Top','Jeans', 'Tie']
plt.stackplot(x,y,labels=labels, edgecolor='black')
plt.legend(loc=2)
```
